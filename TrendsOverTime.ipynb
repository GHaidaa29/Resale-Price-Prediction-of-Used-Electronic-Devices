```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("used_device_data.csv")  

# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values in relevant columns
data = data.dropna(subset=['release_year', 'battery', 'ram'])

# Trend Analysis of Average Battery Capacity and RAM Over the Years
def trends_over_time(data):
    # Group by release year and calculate average battery and RAM
    trends = data.groupby('release_year').agg(
        average_battery=('battery', 'mean'),
        average_ram=('ram', 'mean')
    ).reset_index()

    # Print the trends DataFrame
    print("Trends DataFrame:")
    print(trends)

    # Plotting average battery capacity over the years
    plt.figure(figsize=(12, 6))
    plt.plot(trends['release_year'], trends['average_battery'], marker='o', label='Average Battery Capacity', color='blue')
    plt.title('Trend Analysis of Average Battery Capacity Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Average Battery Capacity (mAh)')
    plt.xticks(trends['release_year'])
    plt.grid()
    plt.legend()
    plt.show()

    # Plotting average RAM over the years
    plt.figure(figsize=(12, 6))
    plt.plot(trends['release_year'], trends['average_ram'], marker='o', label='Average RAM', color='orange')
    plt.title('Trend Analysis of Average RAM Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Average RAM (MB)')
    plt.xticks(trends['release_year'])
    plt.grid()
    plt.legend()
    plt.show()

# Call the trends over time function
trends_over_time(data)
```
