```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("used_device_data.csv")  

# Brand Analysis
def brand_analysis(data):
    # Count the number of devices per brand
    brand_counts = data['brand_name'].value_counts()

    # Plotting the bar graph
    plt.figure(figsize=(12, 6))
    sns.barplot(x=brand_counts.index, y=brand_counts.values, color="skyblue")  # Use color instead of palette
    plt.title('Number of Devices per Brand')
    plt.xlabel('Brand Name')
    plt.ylabel('Number of Devices')
    plt.xticks(rotation=45)
    plt.show()

    # Insights on the most popular brands
    most_popular_brands = brand_counts.head(10)  # Get the top 10 brands
    print("Insights on the Most Popular Brands:")
    print(most_popular_brands)

# Call the brand analysis function
brand_analysis(data)
```
